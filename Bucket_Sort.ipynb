```python
import random

class node():
    def __init__(self,value=None):
        self.next = None
        self.value = value;
    def insert(self,value):
        if self.value == None:
            self.value = value
        elif self.value < value:
            if self.next == None:
                self.next = node(value)
            else:
                self.next.insert(value)
        elif self.value >= value:
            temp = self.value
            self.value = value
            if self.next == None:
                self.next = node(temp)
            else:
                self.next.insert(temp)
    
def bucket_sort(data):
   
    buckets_node = []
    max_data = max(data)
    
    bucket_count = len(data)
    
    bucket_incr = max_data/bucket_count
    
    for i in range(0,bucket_count):
        buckets_node.append(node())
        
    for i in data:
        index = int(i/bucket_incr)
        if i == max_data:
            buckets_node[-1].insert(i)
        else: 
            buckets_node[index].insert(i)
    
    sorted_data = []
    
    for n in buckets_node:
        if n.value != None:
            sorted_data.append(n.value)
            while n.next != None:
                n = n.next
                sorted_data.append(n.value)
    
    print('Sorted Array:')
    print(sorted_data)
                
```

```python
data = []
for i in range(100):
    data.append(random.uniform(0,1))

print('Original Data:')
print(data)
bucket_sort(data)
```
